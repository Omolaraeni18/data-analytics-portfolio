```python
## Imports the pandas library and gives it a shorter alias 'pd' for easier use

import pandas as pd
```

```python
#Import the pyplot module from matplotlib and alias it as 'plt' (used for plotting graphs)

import matplotlib.pyplot as plt
```

```python
##Read the file while skipping blank rows/columns

df = pd.read_excel(r"C:\Users\onite\Downloads\Python_SalesData.xlsx", header=1) 
```

```python
##Shows the first 5 rows of your dataframe by default. Useful for quickly checking what your dataset looks like (columns, data types, sample values)

df.head()
```

```python
##This removes all columns starting with "Unnamed".

df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
##Drop NaN rows

df = df.dropna(how="all")
```

```python
##Shows the first 5 rows of your dataframe by default. Useful for quickly checking what your dataset looks like (columns, data types, sample values). 
##It was used to confirm if all unnamed columns and NaN row dropped

df.head()
```

```python
##Check number of rows and columns

df.shape
```

```python
##General overview of Data

df.info
```

```python
##General overview of Data

df.describe
```

```python
##Ensure proper indexing

df = df.reset_index(drop=True)
```

```python
##Check datatypes...... All datatypes correct but quantity ought to be integer (1,2), but it was float (573.065903), then noticed quantity was  
##misstyped the column ought to be revenue, so the column was changed from quantity to revenue, and quantity was calculated using price and revenue.

print(df.dtypes)
```

```python
## Correct column names.... quantity was renamed to revenue
df = df.rename(columns={"Quantity": "Revenue"})
```

```python
##Check datatypes..... Data types double checked and revenue already changed to float

print(df.dtypes)
```

```python
##Double check first 5 rows and columns and confirm if the correct column name have been effected and quantity correctly replaced with Revenue

df.head()
```

```python
## Round off revenue to 2 decimal place.... revenue was rounded off to two decimal place 

df['Revenue'] = df['Revenue'].round(2)
```

```python
## Calculate true Quantity Sold..... calculated and created a new column for quantity by dividing price from revenue

df['Quantity Sold'] = (df['Revenue'] / df['Price']).round().astype('Int64')
```

```python
##Check no of rows and column.... double checked to confirm a new row has been added
print(df.shape)
```

```python
##Double check first 5 rows and columns and confirm if the new and correct quantity have beeen added to the table
df.head()
```

```python
##Remove leading/trailing spaces

df['Manager'] = df['Manager'].str.strip()
```

```python
##Replace multiple spaces between names with a single space

df['Manager'] = df['Manager'].str.replace(r'\s+', ' ', regex=True)
```

```python
## Verify...Double check if correction has been made and if the multipe spaces have been replaced with a single space

print(df['Manager'].unique()[:10])
```

```python
# Count missing values per column

print(df.isnull().sum())
```

```python
# Check if the entire dataset has any missing values at all

print("Any nulls? →", df.isnull().values.any())
```

```python
# Count duplicates
print("Number of duplicate rows:", df.duplicated().sum())
```

```python
##EDA Sales by purchase type

print(df.groupby('Purchase Type')['Revenue'].sum())
```

```python
##EDA Revenue by day, to find the busiest day of the week

df['Day'] = df['Date'].dt.day_name()
print(df.groupby('Day')['Revenue'].sum().sort_values(ascending=False))
```

```python
## EDA ORDER VALUE DISTRIBUTION...... Helps check if sales are skewed toward small or big transactions.

df['Revenue'].plot(kind='hist', bins=50, title="Distribution of Order Values")
plt.show()
```

```python
## EDA Revenue per manager, per city...

pivot = df.pivot_table(values='Revenue', index='Manager', columns='City', aggfunc='sum').fillna(0)
print(pivot)
```

```python
## Quick overview before answering business questions

print(df.info()) 
```

```python
## Quick overview before answering business questions....  to get all summary statistics for all columns

print(df.describe(include='all'))
```

```python
## Business questions.... 1i.What was the Most Preferred Payment Method?
preferred_payment = df['Payment Method'].value_counts()
print(preferred_payment)
```

```python
## Business questions.... 1ii What was the Most Preferred Payment Method?

most_preferred_payment = preferred_payment.idxmax()
print("Most Preferred Payment Method:", most_preferred_payment)
```

```python
## Business questions.... 2i. Which one was the Most Selling Product by Quantity and by Revenue?
# By Quantity
top_product_qty = df.groupby('Product')['Quantity Sold'].sum().sort_values(ascending=False).head(1)
print("Most Selling Product (Quantity):\n", top_product_qty)
```

```python
## Business questions.... 2ii. Which one was the Most Selling Product by Quantity and by Revenue?
# By Revenue
top_product_revenue = df.groupby('Product')['Revenue'].sum().sort_values(ascending=False).head(1)
print("Most Selling Product (Revenue):\n", top_product_revenue)
```

```python
## Business questions.... 3i. Which City had maximum revenue, and Which Manager earned maximum revenue?
# By Revenue

city_max_revenue = df.groupby('City')['Revenue'].sum().sort_values(ascending=False).head(1)
print("City with Maximum Revenue:\n", city_max_revenue)
```

```python
## Business questions.... 3ii. Which City had maximum revenue, and Which Manager earned maximum revenue?
# By Manager

manager_max_revenue = df.groupby('Manager')['Revenue'].sum().sort_values(ascending=False).head(1)
print("Manager with Maximum Revenue:\n", manager_max_revenue)
```

```python
## Business questions.... 4i. What was the Average Revenue?

df['Month'] = df['Date'].dt.month

avg_nov_dec = df[df['Month'].isin([11,12])]['Revenue'].mean()
print("Average Revenue (Nov & Dec):", avg_nov_dec)
```

```python
## Business questions.... 5. What was the Average Revenue of November & December?

df['Month'] = df['Date'].dt.month
avg_nov_dec = df[df['Month'].isin([11,12])]['Revenue'].mean()
print("Average Revenue (Nov & Dec):", avg_nov_dec)

```

```python
## Business questions.... 6i. What was the Standard Deviation of Revenue and Quantity?
#Revenue

std_revenue = df['Revenue'].std()
print("Standard deviation Revenue:", std_revenue)
```

```python
## Business questions.... 6ii. What was the Standard Deviation of Revenue and Quantity?
#Quantity

std_quantity = df['Quantity Sold'].std()
print("Standard deviation Quantity:", std_quantity)
```

```python
## Business questions.... 7i. What was the Variance of Revenue and Quantity?
#Revenue

var_revenue = df['Revenue'].var()
print("Variance Revenue:", var_revenue)
```

```python
## Business questions.... 7ii. What was the Variance of Revenue and Quantity?
#Quantity

var_quantity = df['Quantity Sold'].var()
print("Variance Quantity:", var_quantity)
```

```python
## Business questions.... 8i. Was the revenue increasing or decreasing over the time?

revenue_trend = df.groupby('Date')['Revenue'].sum()

import matplotlib.pyplot as plt
revenue_trend.plot(figsize=(10,5), title="Revenue Trend Over Time")
plt.ylabel("Revenue")
plt.show()
```

```python
## Business questions.... 8ii. Was the revenue increasing or decreasing over the time?

from scipy.stats import linregress

trend = linregress(range(len(revenue_trend)), revenue_trend.values)
if trend.slope > 0:
    print("Revenue is increasing over time.")
else:
    print("Revenue is decreasing over time.")
```

```python
## Business questions.... 9. What was the Average 'Quantity Sold' & 'Average Revenue' for each product?

avg_product_stats = df.groupby('Product')[['Quantity Sold','Revenue']].mean()
print(avg_product_stats)

```
